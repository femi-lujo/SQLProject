```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT *
        FROM FACILITIES
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """       
  SELECT sub.name, SUM(sub.sessioncost) AS totalrevenue 
  FROM (SELECT f.name, b.memid, 
      		 CASE WHEN b.memid = 0 THEN f.guestcost * b.slots
	 			  ELSE f.membercost * b.slots END AS sessioncost
      FROM Bookings AS b
	  LEFT JOIN Facilities AS f
	  ON b.facid = f.facid
	  LEFT JOIN Members AS m
	  ON b.memid = m.memid) AS sub
  GROUP BY sub.name
  HAVING totalrevenue < 1000;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
        
if __name__ == '__main__':
    main()
 
 
```

```python
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """       
SELECT f.name, m.surname, m.firstname, SUM( b.slots ) AS usagebymember
FROM Bookings AS b
LEFT JOIN Facilities AS f ON b.facid = f.facid
LEFT JOIN Members AS m ON b.memid = m.memid
WHERE b.memid <>0
GROUP BY f.name, m.surname, m.firstname;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
        
if __name__ == '__main__':
    main()
```

```python
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """       
SELECT f.name, m.surname, m.firstname, strftime('%m', b.starttime) AS month, SUM( b.slots ) AS usagepermonth
FROM Bookings AS b
LEFT JOIN Facilities AS f ON b.facid = f.facid
LEFT JOIN Members AS m ON b.memid = m.memid
WHERE b.memid <>0
GROUP BY f.name, m.surname, m.firstname, month;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
        
if __name__ == '__main__':
    main()
```
